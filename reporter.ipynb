```python
import json
import subprocess
import os
import sys
sys.path.append('code')
import experiment
```

```python
def display(path, datadir):
    report_dir = os.path.join(path, datadir)
    metadata = json.load(open(report_dir + '/meta.json'))
    for key in ['task_number', 'base_lr', 'l2_reg', 'mean_pool', 'hidden_dim', 'lstm_hidden_dim', 'wv_dimensions', 'max_epochs']:
        print key, ':', metadata[key]
    experiment.report(report_dir + '/history.cpkl')

def report(subdir):
    path = os.path.join('experiments', subdir)   
    proc = subprocess.Popen(["ls "+ path], stdout=subprocess.PIPE, shell=True)
    (out, err) = proc.communicate()
    for data_dir in out.split():
        display(path, data_dir)
```

```python
report('tn1')
```
